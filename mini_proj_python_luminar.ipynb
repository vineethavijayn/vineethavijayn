```python
#URL Shortener

import string
import random
import os

url_dict = {}
url_dict_1 = {}

def generate_short_url():
    characters = string.ascii_letters + string.digits
    short_url = ''.join(random.choice(characters) for _ in range(7))
    return short_url


def add_shorten_url(original_url):
    short_url = generate_short_url()

    while short_url in url_dict:
        short_url = generate_short_url()
    url_dict[short_url] = original_url

    filename='dictionary.txt'
    with open(filename, 'a') as file:
        for key, value in url_dict.items():
            file.write(f"{key}:{value}\n")


    #print(url_dict)
    return short_url



def read_dict_from_file(filename):
    url_dict_1.clear()
    filename='dictionary.txt'
    with open(filename, 'r') as file:
        for line in file:
            key, value = line.strip().split(':')
            url_dict_1[key] = value
    return url_dict_1




def get_original_url(short_url, filename):
    filename='dictionary.txt'
    with open(filename, 'r') as file:
        for line in file:
            #print(line.strip())
            stored_short_url, long_url = line.strip().split(':')
            if stored_short_url == short_url:
                return long_url
    return None




def get_short_url(original_url, filename):
    filename='dictionary.txt'
    with open(filename, 'r') as file:
        for line in file:
            parts = line.strip().split(':')
            if len(parts) == 2:
                short_url, long_url = parts
                if long_url == original_url:
                    return short_url


def delete_url(url_dict, input_data):
    if input_data in url_dict:
        del url_dict[input_data]
        print(f"Deleted data for '{input_data}'")

    else:
        keys_to_delete = [key for key, value in url_dict.items() if value == input_data]
        if keys_to_delete:
            for key in keys_to_delete:
                del url_dict[key]
            print("Deleted data")
        else:
            print("URL not found")

    return url_dict



def delete_data_from_file(filename, input_data):
    filename='dictionary.txt'
    url_found = False
    with open(filename, 'r') as file_in, open('dictionary1.txt', 'w') as file_out:
        for line in file_in:
            if input_data not in line:
              file_out.write(line)
            else:
                url_found = True


#     file_in.close()
#     file_out.close()
    if url_found:
        print("URL deleted")
        os.remove(filename)
        os.rename('dictionary1.txt', filename)
    else:
        print("URL not found")
        #os.remove(temp_filename)




def check_file_exists(file_path):
    return os.path.exists(file_path)


def is_file_empty(file_path):
    return os.path.getsize(file_path) == 0


def main():

    while True:
        print("\n1. Shorten URL")
        print("2. Retrieve the original URL")
        print("3. Retrive short URL")
        print("4. View the temp dictionary")
        print("5. Delete URL from dictionary")
        print("6. Delete URL from file")
        print("7. View FILE data")
        print("8. Exit")
        choice = input("Enter your choice:")

        if choice == '1':
            original_url = input("Enter the URL to shorten: ")
            short_url = add_shorten_url(original_url)
            print(f"Generated successfully!.. Short URL: {short_url}")


        elif choice == '2':
            short_url = input("Enter the short URL: ")
            original_url = get_original_url(short_url,'dictionary.txt')
            if original_url:
                print(f"Original URL: {original_url}")
            else:
                print("Invalid short URL")


        elif choice == '3':
            original_url = input("Enter the original URL: ")
            short_url = get_short_url(original_url,'dictionary.txt')
            if short_url:
                print(f"Short URL: {short_url}")
            else:
                print(" URL not found ")


        elif choice == '4':
            if url_dict=={}:
                print("No data in dictionary! Press 1 to generate short URL")
            else:
                print(url_dict)


        elif choice=='5':
            input_data=input("Enter the URL to delete")
            del_data=delete_url(url_dict, input_data)


        elif choice=='6':
            input_data=input("Delete URl from FILE")
            filename='dictionary.txt'
            delete_data_from_file(filename, input_data)




        elif choice == '7':
            file_path='dictionary.txt'
            if check_file_exists(file_path):
                loaded_file = read_dict_from_file('dictionary.txt')
                if is_file_empty(file_path):
                    print("The File is Empty")
                else:
                    print("FILE DATA:")
                    print(loaded_file)
            else:
                print("Dictionary/file is empty")



        elif choice == '8':
            print("Exiting.......")
            break
        else:
            print("Invalid choice")

if __name__ == "__main__":
    main()
```
